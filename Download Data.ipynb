Clean Data according to Risk Factors For Chronic Heart Conditions (Please See Paper for additional information)

```python
#Risk Factors include: *High Blood Pressure, *High Cholesterol, Smoking, Diabetes, Obesity, 
#   Inactive Lifestyle, Drinking too Much Alcohol (>1-2 drinks per day), Age (Risk goes up as you age)
# Note * = not avaiable in data
#https://www.cdc.gov/heart-disease/risk-factors/index.html
heart_disease_df = df[['CVDCRHD4','_RFSMOK3','DROCDY4_','_BMI5CAT','_AGE_G','_TOTINDA','DIABETE4']]
```

```python
#Clean Angia or Coronary Heart Disease (This will be the target variable)
heart_disease_df = heart_disease_df[heart_disease_df['CVDCRHD4'] <=3]
heart_disease_df['CVDCRHD4'] = heart_disease_df['CVDCRHD4'].apply(lambda x: 0 if x==1 else 1)
heart_disease_df.rename(columns={'CVDCRHD4':'AngiaOrCoronary'},inplace=True)
```

```python
#Clean the Smoking Group Column, changing format to 0 = not a smoker, 1 = smoker
heart_disease_df = heart_disease_df[heart_disease_df['_RFSMOK3'] < 3]
heart_disease_df['_RFSMOK3'] = heart_disease_df['_RFSMOK3'].apply(lambda x: 0 if x == 1 else 1)
heart_disease_df.rename(columns={'_RFSMOK3':'Smoker'},inplace=True)
```

```python
#Clean Number of Drinks per Day Column
heart_disease_df = heart_disease_df[heart_disease_df['DROCDY4_'] <= 899]
heart_disease_df['DROCDY4_'] = heart_disease_df['DROCDY4_'].dropna()
heart_disease_df.rename(columns={'DROCDY4_':'AlcoholicDrinksPerDay'},inplace=True)
```

```python
#Clean BMI Column to be 0 for Non-Obese and 1 for Obese
heart_disease_df['_BMI5CAT'] = heart_disease_df['_BMI5CAT'].dropna()
heart_disease_df['_BMI5CAT'] = heart_disease_df['_BMI5CAT'].apply(lambda x: 1 if x == 4 else 0)
heart_disease_df.rename(columns={'_BMI5CAT':'Obese'},inplace=True)
```

```python
#Clean Age Column by dropping non-respondents
#1 = 18-24, 2 = 25 to 34, 3 = 35 to 44, 4 = 45 to 54, 5 = 55 to 64, 6 = 65 or older
heart_disease_df.rename(columns={'_AGE_G':'Age'},inplace=True)
```

```python
#Clean Physical Activity Column, 0 for no physical activity, 1 for physical activity within the past month
heart_disease_df = heart_disease_df[heart_disease_df['_TOTINDA'] < 3]
heart_disease_df['_TOTINDA'] = heart_disease_df['_TOTINDA'].apply(lambda x: 0 if x==2 else 1)
heart_disease_df.rename(columns={'_TOTINDA':'Exercise'},inplace=True)
```

```python
#Clean Diabetes Column 0 for anything other then yes, 1 for having diabetes
heart_disease_df = heart_disease_df[heart_disease_df['DIABETE4'] < 5]
heart_disease_df['DIABETE4'] = heart_disease_df['DIABETE4'].apply(lambda x: 1 if x == 1 else 0)
heart_disease_df.rename(columns={'DIABETE4':'Diabetes'},inplace=True)
```
